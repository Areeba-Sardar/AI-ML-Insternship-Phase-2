# 🧠 Auto Tagging Support Tickets Using LLM

### Machine Learning Internship - Task 5

This project demonstrates automatic support ticket classification using Large Language Models (LLMs). The system predicts the most relevant ticket category using Zero-Shot Classification and compares it with a Few-Shot Prompting approach.

## 📚 Import Required Libraries

Import all required Python libraries for data processing, model inference, and evaluation.

```python
import pandas as pd
import numpy as np
from transformers import pipeline
from sklearn.metrics import accuracy_score
```

## 📂 Load Dataset

Load the customer support ticket dataset and inspect its structure.

```python
df = pd.read_csv("data/customer_support_tickets.csv")

print("Shape:", df.shape)

df.head()
```

## 🧹 Data Preprocessing

Select relevant columns and remove missing values.

```python
df = df[
    [
        "Ticket Type",
        "Ticket Subject",
        "Ticket Description"
    ]
]

df = df.dropna()

df.head()
```

## 🔧 Create Ticket Text

Combine ticket subject and ticket description into a single text field for classification.

```python
df["ticket_text"] = (
    df["Ticket Subject"].astype(str)
    + " "
    + df["Ticket Description"].astype(str)
)

df[["Ticket Type","ticket_text"]].head()
```

## 🏷 Candidate Labels

Define ticket categories used for classification.

```python
candidate_labels = [
    "Technical issue",
    "Billing inquiry",
    "Cancellation request",
    "Product inquiry",
    "Refund request"
]
```

## 🤖 Load Zero-Shot Classification Model

Use Facebook BART Large MNLI for support ticket classification.

```python
classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)
```

## 🎯 Predict Top 3 Tags

Generate the three most probable tags for a sample support ticket.

```python
sample_text = df["ticket_text"].iloc[0]

result = classifier(
    sample_text,
    candidate_labels
)

for label, score in zip(
    result["labels"][:3],
    result["scores"][:3]
):
    print(label, round(score,4))
```

## 📊 Model Evaluation

Evaluate the Zero-Shot classifier on a subset of tickets.

```python
sample_df = df.head(20).copy()

predictions = []

for text in sample_df["ticket_text"]:
    
    result = classifier(
        text,
        candidate_labels
    )
    
    predictions.append(
        result["labels"][0]
    )

sample_df["Predicted_Tag"] = predictions

accuracy = accuracy_score(
    sample_df["Ticket Type"],
    sample_df["Predicted_Tag"]
)

print("Accuracy:", accuracy)
```

## 💡 Few-Shot Prompting

Provide example tickets and labels to guide the model before classification.

```python
ticket = sample_df["ticket_text"].iloc[0]

few_shot_prompt = f"""
You are a support ticket classifier.

Examples:

Ticket: App crashes while opening
Tag: Technical issue

Ticket: Payment charged twice
Tag: Billing inquiry

Ticket: Customer wants refund
Tag: Refund request

Ticket: Customer wants to cancel service
Tag: Cancellation request

Ticket: Customer asks product features
Tag: Product inquiry

Now classify this ticket:

{ticket}

Return top 3 probable tags.
"""

print(few_shot_prompt)
```

## ⚖ Comparison of Approaches

Compare Zero-Shot and Few-Shot methods.

```python
comparison = pd.DataFrame({
    "Method":[
        "Zero Shot",
        "Few Shot"
    ],
    "Description":[
        "Classification without examples",
        "Classification using examples"
    ]
})

comparison
```
