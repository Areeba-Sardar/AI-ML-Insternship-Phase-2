# RAG Chatbot Using FAISS & Sentence Transformers

This notebook demonstrates the implementation of a Retrieval-Augmented Generation (RAG) chatbot using Sentence Transformers and FAISS for semantic document retrieval.

## Import Required Libraries

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
```

## Load Documents
The knowledge base is stored in a text file and loaded into memory.

```python
with open("docs.txt", "r", encoding="utf-8") as f:
    documents = f.readlines()

print("Total Documents:", len(documents))
documents[:3]
```

## Generate Embeddings
Convert document text into vector embeddings using Sentence Transformers.

```python
model = SentenceTransformer("all-MiniLM-L6-v2")

embeddings = model.encode(documents)

print("Embedding Shape:", embeddings.shape)
```

## Create FAISS Vector Database
Store embeddings for efficient similarity search.

```python
dimension = embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings))

print("Vector DB created successfully!")
```

## Semantic Search Function
Retrieve the most relevant document for a given query.

```python
def search(query):
    query_embedding = model.encode([query])

    distances, indices = index.search(
        np.array(query_embedding),
        k=1
    )

    return documents[indices[0][0]]
```

## Test the Chatbot

```python
query = "What is Artificial Intelligence?"
answer = search(query)

print("Query:", query)
print("Answer:", answer)
```

## Results

The RAG chatbot successfully retrieves the most relevant document using semantic similarity search. FAISS enables efficient retrieval while Sentence Transformers provide meaningful text embeddings.

## Conclusion

This project demonstrates a simple Retrieval-Augmented Generation (RAG) pipeline using Sentence Transformers and FAISS. The system can retrieve relevant information from a custom knowledge base and provide context-aware responses.
